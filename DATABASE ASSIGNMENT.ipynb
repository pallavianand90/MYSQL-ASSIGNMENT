```python
import mysql.connector as conn
```

```python
mydb = conn.connect(host="localhost", user = "root", passwd = "mysql1234")
cursor = mydb.cursor()
#cursor.execute("create database cardataset")
```

```python
query = "create table cardataset.car(buying VARCHAR(10), maint VARCHAR(10), doors VARCHAR(10), persons VARCHAR(10), lug_boot VARCHAR(10), safety VARCHAR(10), outcome VARCHAR(10))"
cursor.execute(query)
```

```python
import csv
with open("car.data",'r') as f:
    car_data = csv.reader(f , delimiter='\n')
    for i in enumerate(car_data):
        #print(i[1])
        for j in i[1] :
            x = j.split(",")
            res = str(x).replace('[','').replace(']','')
            #print(res)
            #print(f"insert into carda1 values ({str(res)})")
            cursor.execute("insert into cardataset.car values ({values})".format(values=(res)))
mydb.commit()       
```
